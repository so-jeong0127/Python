```python
class Student:
    def __init__(self):
        self.num = 0
        self.name = ''
        self.age = 0
        self.school = ''
        self.cl_num = ''

    def inputData(self):
        self.name = input("이름을 입력하세요. ")
        self.age = eval(input("나이를 입력하세요. "))
        self.school = input("학교이름을 입력하세요. ")
        self.cl_num = input("강좌번호를 입력하세요. ")
        
    def __str__(self):
        return '%5s %10s %7s %10s %10s' % (self.num, self.name, self.age, self.school, self.cl_num)
    def __cmp__(self,other):
        return self.name == other.name
    def __call__(self,other): # 클래스를 함수처럼 사용하고자 할 때 오버라이딩
        st = []
        for student in other:
            std = Student()
            std.num = student[0]
            std.name = student[1]
            std.age = student[2]
            std.school = student[3]
            std.cl_num = student[4]
            st.append(std)
        return st
```
